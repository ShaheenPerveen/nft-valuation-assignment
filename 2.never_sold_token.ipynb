```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import os
import sys
import datetime
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
token_metadata = pd.read_csv('/Users/shaheen/Documents/job_search_jan_2024/home_tasks/nft-valuations/data/token_metadata.csv')
token_metadata.head(5)
```

```python
token_metadata = token_metadata.fillna('Unknown')
token_metadata.nunique()
```

```python
token_metadata['rarest_property_name'] = token_metadata['rarest_property_name'].apply(lambda x: str(x).split(':')[0])
token_metadata.head()
```

```python
# Initialize label encoder
label_encoder = LabelEncoder()

# Encode categorical columns
df_encoded = token_metadata.copy()
for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object':
        df_encoded[col] = label_encoder.fit_transform(df_encoded[col])

# Specify the number of clusters
num_clusters = 5

# Initialize KMeans
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit KMeans to the encoded data
kmeans.fit(df_encoded)

# Get cluster labels
token_metadata['Cluster'] = kmeans.labels_

# View the clusters
print(token_metadata.head())
```

```python
data = pd.read_csv('/Users/shaheen/Documents/job_search_jan_2024/home_tasks/nft-valuations/data/data_with_features.csv')
sold = pd.merge(data, token_metadata[['token_index', 'Cluster']], how='left', on='token_index')
sold.head()
```

```python
avg_valuation = sold.groupby('Cluster').agg({'eth': 'mean'}).reset_index()
avg_valuation
```

```python
never_sold = token_metadata[~token_metadata.token_index.isin(data.token_index)]
never_sold.shape
```

```python
never_sold_evaluation = pd.merge(never_sold[['token_index', 'Cluster']], avg_valuation, how='left', on='Cluster')
never_sold_evaluation.head()
```
